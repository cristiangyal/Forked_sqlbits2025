```python
# ðŸ§¹ Task 1: Filter rows where column 'Drained_after_socializing' = Yes
# Hint: use .filter()
# Filter rows where Drained_after_socializing = Yes
filtered_df = df.filter(df["Drained_after_socializing"] == "Yes")

# Display results
filtered_df.show()
```

```python
# ðŸ”„ Task 2: Select specific columns (e.g., Friends_circle_size and Post_frequency)
# Hint: use .select()
selected_df = df.select('Post_frequency', 'Friends_circle_size')
selected_df.show()
```

```python
# ðŸ§® Task 3: Group by a column (e.g., Personality) and count
# Hint: use .groupBy().count()
grouped_df = df.groupBy('Personality').count().show()
```
