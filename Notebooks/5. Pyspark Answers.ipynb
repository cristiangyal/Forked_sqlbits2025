Here are the solutions to the PySpark exercises.

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySparkDemo').getOrCreate()
```

```python
df = spark.read.option('header', True).csv('/path/to/sample.csv')
```

```python
df.show()
```

```python
filtered_df = df.filter(df['age'] > 30)
filtered_df.show()
```

```python
selected_df = df.select('name', 'age')
selected_df.show()
```

```python
grouped_df = df.groupBy('department').count()
grouped_df.show()
```

```python
# Sample join (assumes df2 exists)
df2 = spark.read.option('header', True).csv('/path/to/other.csv')
joined_df = df.join(df2, df['id'] == df2['id'])
joined_df.show()
```

```python
filtered_df.write.parquet('/path/to/output')
```
