## 1. üîç Initial EDA with Seaborn

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load sample dataset
df = sns.load_dataset("tips")
df.head()
df.shape
````

[Seaborn theme](https://seaborn.pydata.org/tutorial.html)

````python
# Set Seaborn style
sns.set_theme(style="dark", palette='bright')

# Create count plot with region on the y-axis
sns.countplot(data = df, y="sex")

# Show plot
plt.show()
````

````python
# Set Seaborn style
sns.set_theme(style="darkgrid")

# Distribution plot with KDE
sns.histplot(df["total_bill"], kde=True)

````

##### üßÆ What is a KDE?
###### KDE stands for Kernel Density Estimate. 
###### It"s a smoothed curve that estimates the probability density function (PDF) of a continuous variable. 
###### In simpler terms:
###### 
- ###### It shows how the data is distributed, but as a smooth curve instead of jagged bars (like a histogram).
###### 
- ###### Think of it as a more polished way of showing the shape of your data‚Äôs distribution.

````python
# Scatterplot to explore relationships
sns.scatterplot(data=df, x="total_bill", y="tip", hue="time")
plt.show()
````

### üéØ Exercise: Enhance Your Seaborn Plots

###### 1. **Add a title** to each Seaborn plot using `plt.title()`.
###### 2. **Remove axis headers** using `plt.xlabel("")` and `plt.ylabel("")`.



````python

# KDE Histogram with enhancements
plt.figure(figsize=(8, 4))
sns.histplot(df["total_bill"], kde=True)
plt.title("Distribution of Total Bill (with KDE)")
plt.xlabel("Total Bill ¬£")
plt.ylabel("Tip ¬£")
plt.show()


````

````python
# Scatterplot with data labels
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="time")
plt.title("Tips vs Total Bill by Time of Day")
plt.xlabel("")
plt.ylabel("")
plt.show()
````

## 2. üß± Access the Matplotlib Axes from Seaborn

````python
import matplotlib.pyplot as plt

# Create a boxplot and capture the axes
ax = sns.boxplot(data=df, x="day", y="total_bill")
````

## 3. üéØ Refine the Plot with Matplotlib

````python
# Setup the figure and axis
fig, ax = plt.subplots()

# Your plot
sns.boxplot(data=df, x="day", y="total_bill", ax=ax)

# Add titles, labels, annotations, etc.
ax.set_title("Total Bill (¬£) Distribution by Day")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Total Bill (¬£)")
ax.axhline(20, ls="--", color="red", label="¬£20 Reference Line")
ax.legend()

# Layout and export
plt.tight_layout()
plt.savefig('/lakehouse/default/Files/boxplot_totalbill_by_day.png', dpi=300)
#plt.show()
````

_When you use ax.axhline(...) with a label, and then call ax.legend(), the legend will include an entry for the red dashed line labeled "¬£20 Reference Line"._

_This combination is useful for making your plot more informative, especially when you add visual cues like reference lines._

Depending on the environment you are working in, plt.show() can be left out. 
###### 
This is for example the case with Jupyter notebooks, which automatically show all figures created in a code cell.
